```python
import pandas as pd
```

```python
stLink1 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'
pd.read_csv(stLink1)
```

```python
stLink2 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
pd.read_csv(stLink2)
```

```python
stLemonade = pd.read_csv(stLink1)
stBoston = pd.read_csv(stLink2)
```

```python
print(stLemonade.shape)
print(stBoston.shape)
```

```python
print(stLemonade.columns)
print(stBoston.columns)
```

```python
temp = stLemonade['온도']
sale = stLemonade['판매량']
```

```python
import tensorflow as tf
import keras
```

```python
X = tf.keras.layers.Input(shape=[1])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```python
model.fit(temp, sale, epochs = 3000)
```

```python
print("예상값 : ", model.predict([15]))
```
