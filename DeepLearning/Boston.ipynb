```python
import pandas as pd
import tensorflow as tf
```

```python
Boston = pd.read_csv('https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv')
Boston
```

```python
input = Boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat']]
output = Boston[['medv']]
print(input.shape, output.shape)
```

```python
X = tf.keras.Input(shape=[13])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```python
model.fit(input, output, epochs=20)
```

```python
print(model.predict(input[0:5]))
```
